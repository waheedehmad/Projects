```python
documents = data.loc[2:,'Article'].values
```

```python
documents = [str(doc) for doc in documents if isinstance(doc, str)]
topic_model = BERTopic(language="english", low_memory=True ,calculate_probabilities= True,
                     verbose=True)
#can turn probabilities as false for faster\tesr
```

```python
#training the model for the dataset
topics, probs = topic_model.fit_transform(documents)
```

```python
topic_model.get_topic_freq().head(5)
#get important, or frequent topics
```
