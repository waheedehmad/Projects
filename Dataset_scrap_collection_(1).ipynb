```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape article text from the linked page
def get_article_text(full_link):
    linked_page = requests.get(full_link)
    linked_soup = BeautifulSoup(linked_page.text, 'html.parser')
    article_text = linked_soup.find('div', {'class': 'nfl-c-body-part nfl-c-body-part--text'}).text.strip()

    return article_text

# Function to fetch all news articles from the "All News" page, including articles from "Load More" button
def fetch_all_articles():
    all_articles = []

    # Make an HTTP GET request to the "All News" page
    base_url = 'https://www.nfl.com'
    url = base_url + '/news/all-news'
    page = requests.get(url)

    # Parse the HTML of the page
    soup = BeautifulSoup(page.text, 'html.parser')

    # Find all the articles on the page
    articles = soup.find_all('div', {'class': 'd3-l-col__col-12'})

    # Iterate through the articles, follow each link, and scrape data from the linked page
    for article in articles:
        # Extract the link
        link = article.find('a')

        if link is not None and 'href' in link.attrs:
            link_suffix = link['href']
        else:
            continue

        # Concatenate the link suffix with "https://www.nfl.com"
        full_link = base_url + link_suffix

        # Extract the article text from the linked page
        article_text = get_article_text(full_link)

        # Append the data to the list
        all_articles.append({'Article': article_text})

    # Check if there is a "Load More" button
    load_more_button = soup.find('button', {'class': 'nfl-o-cta nfl-o-cta--link'})
    while load_more_button:
        # Extract the link from the "Load More" button
        load_more_link = load_more_button.get('value')

        # Make an HTTP GET request to the next page
        next_page = requests.get(base_url + load_more_link)
        next_soup = BeautifulSoup(next_page.text, 'html.parser')

        # Find all the articles on the next page
        next_articles = next_soup.find_all('div', {'class': 'd3-l-col__col-12'})

        # Iterate through the articles, follow each link, and scrape data from the linked page
        for article in next_articles:
            # Extract the link
            link = article.find('a')

            if link is not None and 'href' in link.attrs:
                link_suffix = link['href']
            else:
                continue

            # Concatenate the link suffix with "https://www.nfl.com"
            full_link = base_url + link_suffix

            # Extract the article text from the linked page
            article_text = get_article_text(full_link)

            # Append the data to the list
            all_articles.append({'Article': article_text})

        # Check if there is another "Load More" button on the next page
        load_more_button = next_soup.find('button', {'class': 'nfl-o-cta nfl-o-cta--link'})

    return all_articles

# Fetch articles from the "All News" page, including articles from "Load More" button
all_articles_data = fetch_all_articles()

# Create a DataFrame to hold the data
df = pd.DataFrame(all_articles_data)

# Save the DataFrame to a CSV file
csv_file = 'nfl_news_articles.csv'
df.to_csv(csv_file, index=False, encoding='utf-8')

print(f"Data has been saved to {csv_file}.")
```
